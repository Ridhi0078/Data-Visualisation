```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn-v0_8")
```

```python
plt.style.use("seaborn-v0_8")
x_coordinates=np.array([0,1,2])*2

plt.bar(x_coordinates-0.25,[10,20,15],width=0.5,label="Current Year",tick_label=["Gold","platinum","Silver"])  # Current year
plt.bar(x_coordinates+0.25,[20,10,12],width=0.5,label="Next Year") # Next year
plt.title("Metal Price Comparsion")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,40)
plt.xlim(-2,5)
plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x_coordinates=np.array([0,1,2])*2

plt.bar(x_coordinates-0.25,[10,20,15],width=0.5,label="Current Year",tick_label=["Gold","platinum","Silver"],color="green")  # Current year
plt.bar(x_coordinates+0.25,[20,10,12],width=0.5,label="Next Year",color="orange") # Next year
plt.title("Metal Price Comparsion")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.show()
```

# Pie Chart

```python
plt.style.use("seaborn-v0_8")
subjects="Maths","Chem","Physics","English"
weightage=[20,10,15,5]

plt.pie(weightage,labels=subjects,explode=(0,0,0.1,0),autopct='%1.1f%%',shadow=True) # explode will slightly shift one part away from pie chart
plt.title("Subjects")
plt.show()
```

# Histogram

```python
Xsn=np.random.randn(100)
sigma=8
u=70
X=np.round(Xsn*sigma+u)
X2=np.round(Xsn*7+40)
print(X,X2)
```

```python
plt.style.use("seaborn-v0_8")

plt.hist(X2,alpha=0.8,label="Physics")
plt.hist(X,alpha=0.9,label="Maths")
plt.ylabel("Prob/FreqCount of Students")
plt.xlabel("Marks Range")
plt.show()
```

 # DATA VISUALISATION CHALLENGE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("movie_metadata.csv")
```

```python
df.head(n=10)
```

```python
df.columns
```

```python
titles=list(df.get('movie_title'))
```

```python
print(type(titles))
```

```python
print(titles[:5])
```

```python
print(titles[0])
```

```python
freq_titles={}

for title in titles:
    length=len(title)

    if freq_titles.get(length) is None:
        freq_titles[length]=1
    else:
        freq_titles[length]+=1
```

```python
print(freq_titles) # for ex there are 138 movies where title length is 7 and so on 
```

```python
X=np.array(list(freq_titles.keys()))
Y=np.array(list(freq_titles.values()))
print(X,Y)
```

```python
plt.scatter(X,Y)
plt.xlabel("Length of movie title")
plt.ylabel("Freq label")
plt.title("Movie Data Visualisation Problem")
plt.show()
```

## Multivariate Normal Distribution

```python
# Bivariate
import numpy as np
mean=np.array([0.0,0.0])
cov=np.array([[0.3,0],[0,2]]) # more spread along y-axis

mean2=np.array([5.0,6.0])
cov2=np.array([[1.3,0.2],[0.2,1.1]])

dist=np.random.multivariate_normal(mean,cov,500)
dist2=np.random.multivariate_normal(mean2,cov2,500)
print(dist.shape)

plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
