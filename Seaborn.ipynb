```python
import seaborn as sns
import numpy as np
```

# BAR PLOT

```python
sns.load_dataset?
```

```python
tips_data=sns.load_dataset('tips')
```

```python
tips_data
```

```python
tips_data[tips_data["sex"]=="Male"]["tip"].mean()
```

```python
sns.barplot(x='sex', y='tip',hue="time", data=tips_data,estimator=np.mean)
```

# COUNT PLOT

```python

sns.countplot(x="sex",data=tips_data)
```

```python
sns.countplot(x="day",hue="sex", data=tips_data)
```

```python
sns.countplot(x="day", data=tips_data)
```

# BOX PLOT

```python
sns.boxplot(x="day",y="tip",hue='sex',data=tips_data)
```

# VIOLIN PLOT

```python
sns.violinplot(x="day",y="tip",hue='sex',data=tips_data)
```

# DISTRIBUTION PLOT

```python
sns.distplot(tips_data["tip"],kde=False) # kernel density estimation
```

# KDE PLOT

```python
sns.kdeplot(tips_data["tip"])
```

# JOINT PLOT

```python
sns.jointplot(x="total_bill",y="tip",data=tips_data,kind="reg")
```

# PAIR PLOT

```python
sns.pairplot(tips_data, hue="sex") # all 3 pairs are plotted against all 3 plots 
```

# HEATMAP

```python
flights=sns.load_dataset('flights')
```

```python
flights
```

```python
flights_pivot=flights.pivot_table(index="month",columns="year",values="passengers")
```

```python
flights_pivot
```

```python
sns.heatmap(flights_pivot)
```

```python
sns.heatmap(flights_pivot,cmap="coolwarm")
```
