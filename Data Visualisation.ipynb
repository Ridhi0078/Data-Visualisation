```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
themes=plt.style.available
print(themes)
```

```python
plt.style.use("seaborn-v0_8")
```

```python
# Line Plot

x=np.arange(10)
y1=x**2
y2=2*x+3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1,color='red',label="Apple",marker='o')
plt.plot(x,y2,color='green',label="Kiwi",linestyle="dashed", marker='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of fruits over time")
plt.legend() # it tells what does red line and green line denote
plt.show()
```

```python
prices=np.array([1,2,3,4])**3
print(prices)
```

```python
plt.plot(prices) # prices will be printed on y and indexes will be printed on x
plt.show()
```

# Scatter plots

```python
# Adjust the size of the plot
plt.figure(figsize=(5,5))
plt.scatter(x,y1,color='red',label="Apple",marker='^')
plt.scatter(x,y2,color='green',label="Kiwi",linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter plot- Prices of fruits over time")
plt.legend() # it tells what does red line and green line denote
plt.show()
```
